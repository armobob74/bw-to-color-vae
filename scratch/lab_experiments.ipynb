```python
from skimage import color
from PIL import Image
import numpy as np
```

```python
img = Image.open("./example_1_original.png")
img_recon = Image.open("example_1_recon.png")
display(img)
display(img_recon)
```

```python
def get_lab(img):
    img_array = np.array(img)
    img_lab = color.rgb2lab(img_array)
    L = img_lab[:,:,0]
    a = img_lab[:,:,1]
    b = img_lab[:,:,2]
    return L,a,b
```

```python
def img_from_lab(L,a,b):
    z =np.zeros((L.shape[0],L.shape[1],3))
    z[:,:,0] = L
    z[:,:,1] = a
    z[:,:,2] = b
    img_rgb = color.lab2rgb(z)
    return Image.fromarray((img_rgb * 255).astype(np.uint8))
```

```python
L,a,b = get_lab(img)
L_recon, a_recon, b_recon = get_lab(img_recon)
```

```python
img_from_lab(L,a,b)
```

```python
img_from_lab(L,a_recon,b_recon)
```
